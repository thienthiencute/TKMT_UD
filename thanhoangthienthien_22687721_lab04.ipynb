### Bài 01

Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng. Bạn hãy: 

a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.

b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm

c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)

```python
import numpy as np
from math import comb

# a. Mô phỏng 10 lần kiểm tra số câu đúng
so_cau = 10
so_lan_thu = 10
xac_suat_dung_1_cau = 1/4  # chọn đúng 1/4 vì 1 trong 4 đáp án đúng

print("a) Kết quả mô phỏng 10 lần kiểm tra:")
for lan in range(so_lan_thu):
    # Sinh ngẫu nhiên số câu đúng trong lần thi này
    cau_dung = np.sum(np.random.rand(so_cau) < xac_suat_dung_1_cau)
    print(f"Lần {lan+1}: {cau_dung} câu đúng")

# b. Tính xác suất để được 4 điểm
# Điểm = 4 * số câu đúng - 2 * số câu sai = 4*X - 2*(10 - X) = 6X - 20
# Muốn được 4 điểm thì: 6X - 20 = 4 => X = 4
x = 4

def binom_pmf(k, n, p):
    return comb(n, k) * (p ** k) * ((1-p) ** (n-k))

xac_suat_4_diem = binom_pmf(x, so_cau, xac_suat_dung_1_cau)
print(f"\nb) Xác suất để sinh viên được đúng 4 câu (4 điểm): {xac_suat_4_diem:.5f}")

# c. Tính E(X) và Var(X)
E_X = so_cau * xac_suat_dung_1_cau
Var_X = so_cau * xac_suat_dung_1_cau * (1 - xac_suat_dung_1_cau)

print(f"\nc) Kỳ vọng E(X) = {E_X}")
print(f"   Phương sai Var(X) = {Var_X}")
```

### Bài 02

Mô phỏng tung một con xúc sắc cân đối đồng chất 5000 lần. Dựa vào giá trị mô phỏng, bạn hãy tìm các giá trị xác suất dưới đây:

a. Xác suất để  số chấm xuất hiện là 4

b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4

```python
import numpy as np

# Số lần tung xúc sắc
so_lan_tung = 5000

# Mô phỏng tung xúc sắc: giá trị từ 1 đến 6
ket_qua = np.random.randint(1, 7, size=so_lan_tung)

# a) Xác suất để số chấm là 4
xac_suat_4 = np.mean(ket_qua == 4)

# b) Xác suất để số chấm >= 4
xac_suat_lon_hon_hoac_bang_4 = np.mean(ket_qua >= 4)

print(f"a) Xác suất số chấm xuất hiện là 4 (ước lượng từ mô phỏng): {xac_suat_4:.4f}")
print(f"b) Xác suất số chấm >= 4 (ước lượng từ mô phỏng): {xac_suat_lon_hon_hoac_bang_4:.4f}")
```

### Bài 03

Tạo 5000 số ngẫu nhiên có phân phối nhị thức với n=50, p=0.7 Bạn Hãy

a. Tìm các giá trị thống kê: min, max, Q1, Q2, Q3

b. Vẽ đồ thị boxplot

c. Vẽ các đồ thị: tần số, tần suất

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Thông số phân phối nhị thức
n = 50
p = 0.7
so_mau = 5000

# a) Tạo 5000 số ngẫu nhiên phân phối nhị thức
data = np.random.binomial(n, p, so_mau)

# Tính các giá trị thống kê
gia_tri_min = np.min(data)
gia_tri_max = np.max(data)
Q1 = np.percentile(data, 25)
Q2 = np.percentile(data, 50)  # trung vị
Q3 = np.percentile(data, 75)

print("Các giá trị thống kê:")
print(f"Min = {gia_tri_min}")
print(f"Q1 (25%) = {Q1}")
print(f"Q2 (Median) = {Q2}")
print(f"Q3 (75%) = {Q3}")
print(f"Max = {gia_tri_max}")

# b) Vẽ boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(data)
plt.title("Boxplot phân phối nhị thức (n=50, p=0.7)")
plt.show()

# c) Vẽ đồ thị tần số và tần suất
plt.figure(figsize=(10, 5))

# Tần số
plt.subplot(1, 2, 1)
sns.histplot(data, bins=range(n+2), kde=False, color='skyblue')
plt.title("Biểu đồ tần số")
plt.xlabel("Giá trị")
plt.ylabel("Tần số")

# Tần suất (xác suất)
plt.subplot(1, 2, 2)
sns.histplot(data, bins=range(n+2), stat="probability", kde=False, color='salmon')
plt.title("Biểu đồ tần suất")
plt.xlabel("Giá trị")
plt.ylabel("Tần suất")

plt.tight_layout()
plt.show()
```

### Bài 04

Một trung tâm bưu điện nhận được 3 cuộc gọi điện thoại mỗi phút. 
Tính xác suất để trung tâm nhận được: 1 cuộc gọi, 2 cuộc gọi, 3 cuộc gọi trong vòng một phút biết số cuộc gọi trong một phút có phân phối Poisson.

```python
import math
import matplotlib.pyplot as plt

# Tham số phân phối Poisson
so_cuoc_goi_trung_binh = 3

# Các giá trị k (số cuộc gọi) từ 0 đến 10
cac_gia_tri_k = list(range(0, 11))

# Tính xác suất theo phân phối Poisson
xac_suat = [(so_cuoc_goi_trung_binh**k) * math.exp(-so_cuoc_goi_trung_binh) / math.factorial(k) for k in cac_gia_tri_k]

# Vẽ biểu đồ cột
plt.figure(figsize=(8,5))
plt.bar(cac_gia_tri_k, xac_suat, color='skyblue', edgecolor='black')
plt.xlabel("Số cuộc gọi (k)")
plt.ylabel("Xác suất P(X=k)")
plt.title("Phân phối Poisson với λ=3")
plt.xticks(cac_gia_tri_k)
plt.show()
```

### Bài 05

Viết chương trình phát sinh ngẫu nhiên mẫu gồm 10.000 phần tử tuân theo phân phối Poisson với tham số λ=4

a. Vẽ biểu đồ histogram để quan sát phân phối mẫu
b. Tính trung bình và phương sai từ dữ liệu mô phỏng, rồi so sánh với giá trị lý thuyết
c. Thực hiện lại câu a với các giá trị λ lần lượt là: 6, 8, 10, 14, 20. Vẽ trên cùng một hình tất cả các đồ thị histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Hàm mô phỏng Poisson và vẽ biểu đồ histogram
def mo_phong_phan_phoi_poisson(so_luong_mau, tham_so_lambda, ten_nhan=None, mau_ve=None):
    # Sinh mẫu Poisson
    mau_poisson = np.random.poisson(lam=tham_so_lambda, size=so_luong_mau)
    # Vẽ histogram
    plt.hist(mau_poisson, bins=range(min(mau_poisson), max(mau_poisson)+2), 
             alpha=0.5, label=ten_nhan, color=mau_ve, edgecolor='black', density=True)
    # Tính trung bình và phương sai mẫu
    trung_binh = np.mean(mau_poisson)
    phuong_sai = np.var(mau_poisson)
    print(f"λ = {tham_so_lambda}: Trung bình mẫu = {trung_binh:.3f}, Phương sai mẫu = {phuong_sai:.3f}")
    print(f"Giá trị lý thuyết: Trung bình = {tham_so_lambda}, Phương sai = {tham_so_lambda}")
    return mau_poisson

# Số lượng mẫu
so_luong_mau = 10000

# a) Mô phỏng lambda=4 và vẽ histogram
print("a) Mô phỏng với λ=4")
plt.figure(figsize=(8,5))
mo_phong_phan_phoi_poisson(so_luong_mau, 4, ten_nhan='λ=4', mau_ve='blue')
plt.xlabel('Giá trị')
plt.ylabel('Tần số chuẩn hóa')
plt.title('Histogram phân phối Poisson với λ=4')
plt.legend()
plt.show()

# c) Mô phỏng và vẽ nhiều histogram với các lambda khác nhau cùng trên 1 hình
danh_sach_lambda = [6, 8, 10, 14, 20]
danh_sach_mau = ['red', 'green', 'orange', 'purple', 'brown']

plt.figure(figsize=(10,6))
for i, lam in enumerate(danh_sach_lambda):
    mo_phong_phan_phoi_poisson(so_luong_mau, lam, ten_nhan=f'λ={lam}', mau_ve=danh_sach_mau[i])

plt.xlabel('Giá trị')
plt.ylabel('Tần số chuẩn hóa')
plt.title('Histogram phân phối Poisson với các λ khác nhau')
plt.legend()
plt.show()
```
